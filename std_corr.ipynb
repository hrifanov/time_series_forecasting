```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Given correlation results
data = {
    'NHiTS 5min': [0.518,  0.316],
    'NHiTS 15min': [0.451,  0.236],
    'TFT 5min': [0.432,  0.253],
    'TFT 15min': [0.435,  0.295]
}

# Create a DataFrame
df = pd.DataFrame(data, index=['Active vs Passive Portfolio Difference', 'Active Portfolio Profit'])

# Plot the correlation matrix
plt.figure(figsize=(10,  3))
sns.heatmap(df, annot=True, cmap='coolwarm', fmt='.3f', cbar=True,  xticklabels=df.columns, yticklabels=df.index)
plt.title('Correlation Matrix of Active Portfolio Metrics with Stock Standard Deviation')
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Given correlation results
data = {
    'nhits   5': [0.518,   0.316],
    'nhits   15': [0.451,   0.236],
    'tft   15': [0.414,   0.253],
    'tft   5': [0.415,   0.295]
}

# Create a DataFrame
df = pd.DataFrame(data, index=['Active Portfolio Difference vs. Stock Standard Deviation', 'Active Portfolio Profit vs. Stock Standard Deviation'])

# Create subplots
fig, axs = plt.subplots(4,   1, figsize=(6,   12))

# Plot each subplot
for i, (timeframe, row_data) in enumerate(data.items()):
    axs[i].set_title(f'{timeframe} vs. Standard Deviation')
    sns.heatmap(pd.DataFrame(row_data, index=df.index, columns=[timeframe]), annot=True, cmap='coolwarm', fmt='.3f', cbar=True, ax=axs[i])
    axs[i].set_xlabel('Metric')
    axs[i].set_ylabel('')

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```
